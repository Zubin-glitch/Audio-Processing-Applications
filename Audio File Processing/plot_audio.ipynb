```python
import wave
import matplotlib.pyplot as plt
import numpy as np
```

```python
audio_obj = wave.open("LJ025-0076.wav")
print(audio_obj.getparams())
sample_freq = audio_obj.getframerate()
n_samples = audio_obj.getnframes()
signal_wave_array = audio_obj.readframes(-1) #read all frames = -1 in params of readframes method

audio_obj.close()
```

```python

audio_time = n_samples/sample_freq
print(audio_time)
signal_array = np.frombuffer(signal_wave_array, dtype=np.int16)
times = np.linspace(0, audio_time, num=n_samples)
```

```python
# Plotting the graph to show the audio signal
plt.figure(figsize=(15, 5))
plt.plot(times, signal_array)
plt.title("Audio Signal Analysis")
plt.xlabel("Time (seconds)")
plt.ylabel("Signal Waveform")
plt.xlim(0, audio_time)
plt.show()
```
